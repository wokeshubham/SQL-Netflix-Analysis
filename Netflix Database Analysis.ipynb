```python
import pandas as pd 
import pandasql as ps 
```

```python
netflix_dataset = pd.read_csv("netflixdata.csv")
print(netflix_dataset.shape)
```

```python
ps.sqldf("""

    SELECT * FROM netflix_dataset as df

""")
```

**Client asked to pull all the Movies made in 2015.**

```python
ps.sqldf("""

    SELECT * FROM netflix_dataset as df
    WHERE type = 'Movie'  AND
          release_year = 2015

""")
```

**Client asked the title of all TV Shows that are made in India after 2016**

```python
ps.sqldf("""

    SELECT title FROM netflix_dataset as df
    WHERE country = 'India' AND
    release_year > 2015    

""")
```

**Client asked all Movies with the TV rating of R that was made in US**

```python
ps.sqldf("""

    SELECT * FROM netflix_dataset as df
    WHERE rating = 'R' AND
    country = 'United States'

""")
```

**Client asked for cast of Movie= 21 & over and the genre of the Movie**

```python
ps.sqldf("""

    SELECT df.title, df.cast, df.listed_in FROM netflix_dataset as df
    WHERE title = '21 & Over'

""")
```

**Client asked to pull all the Movies/ TV show made by the director of the above movie.**

```python
ps.sqldf("""

    SELECT * from netflix_dataset as df
    WHERE df.director IN(

        SELECT df.director FROM netflix_dataset as df
        WHERE df.title = '21 & Over'

    )

""")
```

**Client asked for 10 latest shows/movies that was added.**

```python
ps.sqldf("""

    SELECT * FROM netflix_dataset as df
    ORDER BY date_added DESC
    LIMIT 10

""")
```

**Client asked for all Indian movies that start with the letter 'S' and released after 2018.**

```python
ps.sqldf("""

    SELECT df.title, df.release_year FROM netflix_dataset as df
    WHERE type = 'Movie'  AND
    country = 'India' AND
    title LIKE 'S%'  AND
    release_year > 2018
    ORDER BY release_year DESC

""")
```

**Client asked for the movie with the description 'A scrappy but poor boy worms his way into...' Display only the movie name and the release year**

```python
ps.sqldf("""

    SELECT df.title, df.release_year FROM netflix_dataset as df
    WHERE description LIKE 'A scrappy but poor boy worms his way into%' 

""")
```

**Client asked me to list 10 earliest Si-fi, international Movies**

```python
ps.sqldf("""

    SELECT * FROM netflix_dataset as df
    WHERE listed_in LIKE '%Sci-Fi%' AND
    listed_in LIKE '%International Movies%'
    ORDER BY release_year ASC
    LIMIT 10

""")
```

**Client asked me to list all the movies or tv shows that were added on November 2019**

```python
ps.sqldf("""

    SELECT * FROM netflix_dataset as df
    WHERE date_added LIKE 'November%' AND
    date_added LIKE '%2019'
    ORDER BY date_added

""")
```

**Client asked the list of latest Horror Movies and where they where made**

```python
ps.sqldf("""

    SELECT df.title, df.country FROM netflix_dataset as df
    WHERE listed_in LIKE '%Horror Movie%'
    ORDER BY release_year DESC
    LIMIT 10

""")
```
